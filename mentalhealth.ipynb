```python
# =============================================
# AnÃ¡lisis de datos: Global Suicide, Mental Health, Substance Use Disorders Trends
# Incluye limpieza automÃ¡tica, regresiÃ³n y clasificaciÃ³n
# =============================================

# Importar librerÃ­as necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold, StratifiedKFold, train_test_split, cross_val_score, cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, f1_score, classification_report
from sklearn.decomposition import PCA

#  Cargar el dataset
file_path = "/content/crude suicide rates.csv"  # <- asegÃºrate de subirlo antes de correr
df = pd.read_csv(file_path, encoding='utf-8', low_memory=False)
print("Dimensiones iniciales:", df.shape)
df.head()

#  Limpieza automÃ¡tica: convertir valores numÃ©ricos en texto a float
df_num = df.copy()
for col in df_num.columns:
    df_num[col] = pd.to_numeric(df_num[col].astype(str).str.extract(r'(-?\d+\.?\d*)')[0], errors='coerce')

df_num = df_num.dropna(axis=1, how='all')
df_num = df_num.fillna(df_num.median(numeric_only=True))

print("Dimensiones del dataset numÃ©rico:", df_num.shape)
display(df_num.head())

#  Identificar variable objetivo (target)
candidates = [c for c in df_num.columns if ('suicide' in c.lower()) or ('crude' in c.lower()) or ('rate' in c.lower())]
target_col = candidates[0] if candidates else df_num.columns[-1]
print("Target elegido:", target_col)

#  Separar X e y
X = df_num.drop(columns=[target_col])
y = df_num[target_col].values

# Escalamiento
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# =============================================
# ðŸ”¹ REGRESIÃ“N
# =============================================
kf = KFold(n_splits=5, shuffle=True, random_state=42)
lr = LinearRegression()
rf_reg = RandomForestRegressor(n_estimators=200, random_state=42)

r2_lr = cross_val_score(lr, X_scaled, y, cv=kf, scoring='r2')
r2_rf = cross_val_score(rf_reg, X_scaled, y, cv=kf, scoring='r2')

pred_lr = cross_val_predict(lr, X_scaled, y, cv=kf)
pred_rf = cross_val_predict(rf_reg, X_scaled, y, cv=kf)

rmse_lr = np.sqrt(mean_squared_error(y, pred_lr))
rmse_rf = np.sqrt(mean_squared_error(y, pred_rf))

print("\n REGRESIÃ“N:")
print(f"LinearRegression RÂ²: {r2_lr.mean():.4f} Â± {r2_lr.std():.4f}")
print(f"RandomForestRegressor RÂ²: {r2_rf.mean():.4f} Â± {r2_rf.std():.4f}")
print(f"LinearRegression RMSE: {rmse_lr:.4f}")
print(f"RandomForestRegressor RMSE: {rmse_rf:.4f}")

# Importancias RF
rf_reg.fit(X_scaled, y)
importances_reg = pd.Series(rf_reg.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 10 variables (RegresiÃ³n - RF):")
display(importances_reg.head(10))

# GrÃ¡fica de regresiÃ³n
plt.figure(figsize=(6,5))
plt.scatter(y, pred_rf, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel('Valor real')
plt.ylabel('Predicho (RF)')
plt.title('PredicciÃ³n vs Real (RandomForestRegressor)')
plt.show()

# =============================================
# ðŸ”¹ CLASIFICACIÃ“N
# =============================================
y_class = pd.qcut(y, q=3, labels=["bajo","medio","alto"])
print("\nDistribuciÃ³n de clases:")
print(y_class.value_counts())

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
clf_rf = RandomForestClassifier(n_estimators=200, random_state=42)
clf_lr = LogisticRegression(max_iter=1000, multi_class='multinomial', solver='lbfgs')

acc_rf = cross_val_score(clf_rf, X_scaled, y_class, cv=skf, scoring='accuracy')
acc_lr = cross_val_score(clf_lr, X_scaled, y_class, cv=skf, scoring='accuracy')

print("\n CLASIFICACIÃ“N (CV):")
print(f"RandomForestClassifier accuracy: {acc_rf.mean():.4f} Â± {acc_rf.std():.4f}")
print(f"LogisticRegression accuracy: {acc_lr.mean():.4f} Â± {acc_lr.std():.4f}")

# Entrenar para test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_class, test_size=0.2, random_state=42, stratify=y_class)
clf_rf.fit(X_train, y_train)
clf_lr.fit(X_train, y_train)

pred_rf_test = clf_rf.predict(X_test)
pred_lr_test = clf_lr.predict(X_test)

print("\nðŸ”¸ RandomForest - Test:")
print("Accuracy:", accuracy_score(y_test, pred_rf_test))
print("F1 macro:", f1_score(y_test, pred_rf_test, average='macro'))
print(classification_report(y_test, pred_rf_test))

print("\nðŸ”¸ LogisticRegression - Test:")
print("Accuracy:", accuracy_score(y_test, pred_lr_test))
print("F1 macro:", f1_score(y_test, pred_lr_test, average='macro'))
print(classification_report(y_test, pred_lr_test))

# Importancias clasificaciÃ³n
importances_clf = pd.Series(clf_rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nTop 10 variables (ClasificaciÃ³n - RF):")
display(importances_clf.head(10))

# GrÃ¡fica de importancias
plt.figure(figsize=(8,4))
importances_clf.head(10).plot(kind='bar')
plt.title('Top 10 Importancias (ClasificaciÃ³n - RandomForest)')
plt.ylabel('Importancia')
plt.show()

# PCA para visualizaciÃ³n
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(6,5))
labels = np.array(y_class)
for lab in np.unique(labels):
    plt.scatter(X_pca[labels==lab,0], X_pca[labels==lab,1], label=lab, alpha=0.6)
plt.title('VisualizaciÃ³n PCA (tertiles del target)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()

print("\n AnÃ¡lisis completado.")
```
